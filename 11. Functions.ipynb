### Function for getting EVEN Numbers in a List of 1-26

```python
rock=list(range(1,26))
def even(x):
    if x%2==0:
        return x

list(filter(even,rock))
```

### writing above code Using LAMBDA

```python
even1=list(range(1,31))
list(filter(lambda a: a%2==0,even1))
```

### Function for getting ODD Numbers in a List of 1-26

```python
sizer=list(range(1,26))
def odd(y):
    if y%2!=0:
        return y
list(filter(odd,sizer))
```

### writing above code Using LAMBDA

```python
odd_num=list(range(1,26))
list(filter(lambda a: a%2!=0, odd_num))
```

## Recursion Function
The Function call itself recursion Function

```python
def nazi(): # we are creating a function caklled nazi
    print("nzzrr") # telling it to print "nzzrr"
    nazi()        # then tell it to run the function same function
nazi()        # so it runs the function, inside function we asked to run the function,
# it repeats like a loop, but its not a loop, becoz it is a recusrion
# And recusrion has Limit on how many times it should be running
```

### Finding the Recursion Limit of Python

```python
import sys  # we imported python's System Library
sys.getrecursionlimit()  # see down below the recursion limit is 3000
```

### Increasing the Recursion Limit

```python
sys.setrecursionlimit(5000) # after setting this limit will change
sys.getrecursionlimit()  # Now recursion limit is 5000
```

### Limiting The recursion without decreasing the limit

```python
def hello(c):  # we defined a function, and c is the parameter
    if c==0:   # we are checking if condition for "if c becomes 0"
        return "Recursion Stopped" # when if condition becomes true it returns the statement
    print("bot",c) 
    return hello(c-1) # In function, we're telling it to run the same function, and "c-1" as parameter
hello(10) # function calling with "10" as parameter
```

## Factorial Number with Recursion Function

```python
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)  
user_num=int(input("enter your factorial number: "))
factorial(user_num)
```
